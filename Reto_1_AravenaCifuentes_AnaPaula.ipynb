## Torres de Hanoi con Divide y vencerás

Resuelve el problema de las [Torres de Hanoi](https://es.wikipedia.org/wiki/Torres_de_Han%C3%B3i) para un número arbitrario de fichas utilizando la técnica de divide y vencerás.

```python
# Torres de Hanoi
def hanoi(n, origen, destino, auxiliar):
  if n==1:
    print(f"Mover disco 1 de {origen} a {destino}")
    return
  hanoi(n-1, origen, auxiliar, destino)
  print(f"Mover disco {n} de {origen} a {destino}")
  hanoi(n-1, auxiliar, destino, origen)

def resolver_hanoi():
  origen=input("Origen:")
  destino=input("Destino:")
  auxiliar=input("Auxiliar:")
  n=int(input("N° de discos:"))
  print(f"Resolviendo las Torres de Hanoi con {n} discos:")
  hanoi(n, origen, destino, auxiliar)

resolver_hanoi()
```

## Sucesión de Fibonacci

Cálcula el n-ésimo término de la [Sucesión de Fibonacci](https://es.wikipedia.org/wiki/Sucesi%C3%B3n_de_Fibonacci) mediante un algoritmo recursivo y otro iterativo. Representa gráficamente cómo crece el tiempo de cómputo en función del número de términos para ambos algoritmos.

```python
#Sucesión_de_Fibonacci

def fib_iterativo(n):
  fib = [0, 1]
  for i in range(2, n + 1):
    fib.append(fib[i - 1] + fib[i - 2])
  return fib[n]

fib_iterativo(1)
```

```python
fib_iterativo(11)
```

```python
def fib_recursivo(n):
  if n <= 1:
    return n
  else:
    return fib_recursivo(n - 1) + fib_recursivo(n - 2)
fib_recursivo(1)
```

```python
fib_recursivo(11)
```

```python
import timeit
import matplotlib.pyplot as plt

n_valores = list(range(1, 31))

t_recursivos = [timeit.timeit(lambda: fib_recursivo(n), number=1) for n in n_valores]
t_iterativos = [timeit.timeit(lambda: fib_iterativo(n), number=1) for n in n_valores]

plt.plot(n_valores, t_recursivos, label="Recursivo")
plt.plot(n_valores, t_iterativos, label="Iterativo")
plt.xlabel("N° de términos de Fibonacci (n)")
plt.ylabel("Tiempo de ejecución (s)")
plt.title("Comparación de tiempos de ejecución")
plt.legend()
plt.show()
```

## Devolución de cambio por técnica voraz

Resuelve el [Problema del Cambio de Moneda](https://es.wikipedia.org/wiki/Problema_de_cambio_de_monedas) utilizando una técnica voraz.

```python
import math
def cambio_voraz(monto, denominaciones):
  denominaciones = sorted(denominaciones, reverse=True)
  cambio = []
  if monto<=0:
    print("El monto",monto, "es negativo, por lo que se calculará el monto considerando su valor absoluto.")
    monto=abs(monto)
  if isinstance(monto, float):
    print("El monto",monto, "es un decimal, por lo que se calculará el monto truncando su valor.")
    monto=math.trunc(monto)
  for d in denominaciones:
    while monto >= d:
      cambio.append(d)
      monto -= d
  return cambio

devolver = 70350
denom = [50000, 20000, 10000, 5000, 2000, 1000, 500, 100, 50, 10, 5]

cambio_dar = cambio_voraz(devolver, denom)
print("Monto a devolver:", devolver)
print("Cambio resultante:", cambio_dar)
```

## N-Reinas por técnica de vuelta atrás
Resuelve el [Problema de las N-Reinas](https://es.wikipedia.org/wiki/Problema_de_las_ocho_reinas) en un tablero de dimensión N mediante la técnica de la vuelta atrás (backtraking).

```python
# Problema de las N-Reinas

def se_puede(tablero, fila, columna, n):
  for i in range(fila):
    if tablero[i] == columna or tablero[i] - i == columna - fila or tablero[i] + i == columna + fila:
      return False
  return True

def ver_tablero(tablero):
  for fila in tablero:
    line = ['Q' if i == fila else '.' for i in range(len(tablero))]
    print(' '.join(line))
  print()

def resolver_n_reinas_util(tablero, fila, n):
  if fila == n:
    ver_tablero(tablero)
    return True
  resultado = False
  for columna in range(n):
    if se_puede(tablero, fila, columna, n):
      tablero[fila] = columna
      resultado = resolver_n_reinas_util(tablero, fila + 1, n) or resultado
      tablero[fila] = -1
  return resultado

def resolver_n_reinas(n):
  tablero = [-1] * n
  if not resolver_n_reinas_util(tablero, 0, n):
    print("No hay respuesta para la dimension", n,"x",n)

resolver_n_reinas(4)
```
